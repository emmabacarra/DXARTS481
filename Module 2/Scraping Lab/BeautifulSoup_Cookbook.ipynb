```python
import os #to do system stuff, like making folders

# import the requests library
import requests

# import beautifulsoup
from bs4 import BeautifulSoup

#import pypdf
import pypdf

#import reportlab to create pdfs
from reportlab.pdfgen import canvas
```

```python
# get the first page of a website
webp = requests.get('https://www.nytimes.com/section/education')

# create a beautifulsoup object using the html
soup = BeautifulSoup(webp.text)
```

# Search filters in  BeautifulSoup - Useful to find tags, classes, kewords...

```python
#Search a string: Beautiful Soup will perform a tag-name match against that exact string. This code finds all the <div> tags in the document:
soup.find_all('div')
```

```python
#using a regular expression: filter against that regular expression using its search() method. 
#This code finds all the tags whose names start with the letter "a"
import re
for tag in soup.find_all(re.compile("^a")):
    print(tag.name)
```

```python
#This code finds all the tags whose names contain the letter 't':

for tag in soup.find_all(re.compile("t")):
    print(tag.name)
# html
# title
```

```python
#The value True matches every tag it can. 
#This code finds all the tags in the document, but none of the text strings:

for tag in soup.find_all(True):
    print(tag.name)
```

```python
#Here's a function that returns True if a tag defines the "class" attribute but doesn't define the "id" attribute:

def has_class_but_no_id(tag):
    return tag.has_attr('class') and not tag.has_attr('id')
```

```python
soup.find_all(has_class_but_no_id)
```

```python
# Find all tags and extract their IDs
for tag in soup.find_all(True):
    if tag.has_attr('id'):
        print(tag.name, 'id: ', tag['id'])
```

```python
# Find all tags and extract their class:
for tag in soup.find_all(True):
    if tag.has_attr('class'):
        print('tag: ', tag.name, 'class: ', tag['class'])
```

```python
# Find all tags with a class attribute and count the class occurrences

from collections import Counter

class_counter = Counter()

for tag in soup.find_all(True):
    if tag.has_attr('class'):
        for class_name in tag['class']:
            class_counter.update([(tag.name, class_name)])

# Sort the classes by their counts in descending order and print the results
sorted_classes = sorted(class_counter.items(), key=lambda item: item[1], reverse=True)

for (tag_name, class_name), count in sorted_classes:
    print("tag: ", tag_name, "/  class name: ", class_name, " /  repeated: ", count)
```

```python
# This code finds all the <a> tags withh the <css-1u3p7j1> class:
# Using class as a keyword argument will give you a syntax error. 
# As of Beautiful Soup 4.1.2, you can search by CSS class using the keyword argument class_:

for tag in soup.find_all("a", class_="css-1u3p7j1"):
    print(tag.text)
```

```python
#This code finds all the <li> tags withh the <eb97p613> class:

for tag in soup.find_all("li", class_="eb97p613" ):
    print(tag.text)
```

```python
# find all tags that have a href class and print the tag name, other classes + the link in href 

for tag in soup.find_all(href=True):
    class_name = tag.get('class')
    print('Tag: ', tag.name, "Href: ",  tag["href"], "Class: ",  class_name)
```

### You can nest some of these searcher by adding it all within the parenthesis, like the following examples:

```python
#find all the sections, then the divs, then the p tags.
titles = soup.select('section div p')

for t in titles:
    print(t.text)
```

```python
# find the main tag, then look inside and find all the divs, and then inside the divs find all the a tags:

titles = soup.select('main div a')

for t in titles:
    print(t.text)
```

# we can try to get all the links from the ny times, education section, first page:

```python
# Get all the links under the section tag, then under the div tag and under the tag a with class css-1u3p7j1 

titles = soup.select('section div a.css-1u3p7j1')

#then iterate through the results, find the href part and to all the hrefs found, we will concatenate the https part so that we can actually
#obrain a link. 
#then we just print the links

for t in titles:
    href = t.get('href')
    if href.startswith("/"):
        href = "https://www.nytimes.com/" + href
    print(href)
```
