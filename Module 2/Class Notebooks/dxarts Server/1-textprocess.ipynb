```python
import re
import spacy
from tqdm.notebook import tqdm 

nlp = spacy.load('en_core_web_md')

input_file = "neuromancer.txt"

sentences = []
with open(input_file, 'r', encoding='utf-8') as f:
    text = f.read()

    text = re.sub(r'(?<=\S)-\n', '', text)  # -\n
    
    text = text.replace('\n', '')

    for sentence in tqdm(nlp(text).sents, desc="Processing sentences", unit="sentence"):
        sentences.append(sentence.text.strip()) # spacy.tokens.span.Span
```

```python
import json
import random
# create jsonl
data = []
for i in range(len(sentences) - 1):
    input_sentence = sentences[i]
    output_sentence = sentences[i + 1]
    data.append({"input": input_sentence, "output": output_sentence})

# shuffle
random.shuffle(data)

# 80% training，20% validation
split_index = int(0.8 * len(data))
train_data = data[:split_index]
eval_data = data[split_index:]

# 80% training
with open("notes.jsonl", "w", encoding="utf-8") as f:
    for item in train_data:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")

# 20% validation
with open("notes_validation.jsonl", "w", encoding="utf-8") as f:
    for item in eval_data:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")
```
