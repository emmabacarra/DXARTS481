#### calculating frame intervals

```python
lookback_times = []
n=1
for file in maps_files:
    with fits.open(file) as manga_map:
            
        primary_header = manga_map['PRIMARY'].header
        plateifu = primary_header.get("PLATEIFU", None)
        scinpvel = primary_header.get("SCINPVEL", None)

        if scinpvel is not None:
            # redshift z = v / c
            z = scinpvel / const.c.to('km/s').value
            lookback_time = cosmo.lookback_time(z).to('Gyr').value
            lookback_times.append(lookback_time)
        
        print(f'{n}/{len(maps_files)}', end='\r')
        n+=1
```

```python
fps = 60
lookbacktimes = np.unique(lookback_times)
n_lookbacktimes = len(lookbacktimes)
print(f'[Unique Years]: {n_lookbacktimes}')

# diffs = [b - a for a, b in zip(lookbacktimes, lookbacktimes[1:])]
diffs = np.diff(lookbacktimes) # same as above, but more efficient
print(f'[Diffs] | Mean: {np.mean(diffs):.6f} Gyr | Q1: {np.percentile(diffs, 25):.6f} Gyr | Median: {np.median(diffs):.6f} Gyr | Q3: {np.percentile(diffs, 75):.6f} Gyr')

percentile = 90
step_size = np.percentile(diffs, percentile)
steps = np.arange(min(lookbacktimes), max(lookbacktimes), step_size)
frames = len(steps)

vid_length = frames / fps
m, s = divmod(vid_length, 60)
h, m = divmod(m, 60)

print(f'[Range]: {range:.3f} Gyr')
print(f'[Step size ({percentile}th percentile)]: {step_size} Gyr ({step_size * 1e9:.2f} years)')
print(f'[Video length]: {h:.0f}h {m:.0f}m {s:.2f}s ({int(frames)} frames)')
```

```python
# check how many lookback_times are in each step
step_counts = np.zeros(len(steps))
for i, step in enumerate(steps):
    step_counts[i] = len([1 for t in lookback_times if t >= step and t < step + step_size])

# find where step_counts is 0
zero_steps = np.where(step_counts == 0)[0] # [0] for True
print(f'[Zero steps]: {len(zero_steps)}')

plt.figure(figsize=(10, 5));
plt.plot(steps, step_counts);
plt.xlabel(f'Time Steps ({step_size:.6f} Gyr)');
plt.ylabel('Files');
```

### point clouds (relative position)

```python
def generate_point_cloud(file, depth_scale_factor = 30, spread_factor = 0.02, depth_layers = 100):
    with fits.open(file) as hdul:
        
        data = hdul['EMLINE_GFLUX'].data
        iso_forest = IsolationForest(contamination=0.5, random_state=42)  # 20% expected meaningful channels
        channels = [(shannon_entropy(data[i]), np.std(data[i])) for i in range(35)]
        labels = iso_forest.fit_predict(channels)
        selected_channels = np.where(labels == 1)[0]
        
        emission_flux = np.sum(hdul['EMLINE_GFLUX'].data[selected_channels], axis=0)
        g_band_flux = hdul['SPX_MFLUX'].data
        
        flux_map = emission_flux + g_band_flux # brightness
        flux_map = np.where(flux_map == 0, np.nan, flux_map)
        flux_map = np.log1p(flux_map) 
        vel_map = hdul['EMLINE_GVEL'].data[23] # doppler shift for "depth"
        disp_map = hdul['EMLINE_GSIGMA'].data[23] # for galaxy thickness

    v_max = np.nanmax(np.abs(vel_map))
    depth_map = (vel_map / v_max) * depth_scale_factor

    # preview flux and depth maps
    # fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    # ax[0].imshow(flux_map, cmap='inferno'), ax[0].axis('off')
    # ax[1].imshow(depth_map, cmap='inferno'), ax[1].axis('off')

    # getting rid of unwanted values
    valid_mask = np.isfinite(flux_map) & np.isfinite(depth_map)
    flux_map = flux_map[valid_mask]
    depth_map = depth_map[valid_mask]

    # x, y, z coordinates (NOT the real coordinates in space)
    y, x = np.indices(vel_map.shape)
    x, y, z = x[valid_mask], y[valid_mask], depth_map
    print(f'x: {x.shape} | y: {y.shape} | z: {z.shape}')

    flux_norm = (flux_map - np.min(flux_map)) / (np.max(flux_map) - np.min(flux_map))
    # colors = np.c_[flux_norm, flux_norm, flux_norm]  # grayscale mapping
    colors = cm.managua_r(flux_norm)[:, :3] 
    # colors = flux_norm

    # point cloud!! ---------------------------------------------------------
    disp_norm = (disp_map - np.nanmin(disp_map)) / (np.nanmax(disp_map) - np.nanmin(disp_map))
    disp_spread = disp_norm * spread_factor

    # adding more depth
    all_points = []
    for i in range(depth_layers):
        z_offset = (i - depth_layers // 2) * disp_spread[valid_mask] # adjustable
        new_z = z + z_offset
        all_points.append(np.c_[x, y, new_z])

    all_points = np.vstack(all_points)
    all_colors = np.tile(colors, (depth_layers, 1))

    pcd_disp = o3d.geometry.PointCloud()
    pcd_disp.points = o3d.utility.Vector3dVector(all_points)
    pcd_disp.colors = o3d.utility.Vector3dVector(all_colors)
    # voluminous pixel
    voxel_grid = o3d.geometry.VoxelGrid.create_from_point_cloud(pcd_disp, voxel_size=1)
    
    return voxel_grid

# data_folder = '../../MaNGA/raw'
# maps_files = [os.path.join(data_folder, f) for f in os.listdir(data_folder)]

# selected_plateifus = ['10001-12701', '10001-9101', '10148-12704']
# files = [next(f for f in maps_files if plateifu in f) for plateifu in selected_plateifus]
# voxel_grids = [generate_point_cloud(file) for file in files]
# o3d.visualization.draw_geometries(voxel_grids)

plateifu = '10001-12701'
file = next(f for f in maps_files if plateifu in f)
voxel_grid = generate_point_cloud(file,
                                  depth_scale_factor = 30, 
                                  spread_factor = 0.02, 
                                  depth_layers = 100
                                  )
o3d.visualization.draw_geometries([voxel_grid])
```

### pca clustering

```python
# Load data files
data_folder = '../../MaNGA/raw'
maps_files = [os.path.join(data_folder, f) for f in os.listdir(data_folder)]

# Define life cycle stages for galaxies
life_cycle_stages = {
    'early formation': [0, 1, 2, 3, 4, 5, 8],
    'star formation': [7, 9, 10, 14, 19, 23, 24, 27, 28, 31, 33],
    'stellar evolution': [11, 12, 13, 22, 25, 26],
    'agn activity': [15, 16, 17, 20, 21],
    'late stage evolution': [18, 19, 25, 26, 30, 32, 34],
    'passive evolution': [11, 12, 13, 18, 19, 20, 21, 25, 26]
}

def generate_point_cloud(file_path, colormap=cm.managua_r, depth_scale_factor=30, spread_factor=0.5, depth_layers=100, print_info=False):
    kms = u.km / u.s

    with fits.open(file_path) as hdul:
        # Object info
        primary_header = hdul['PRIMARY'].header
        scinpvel = primary_header.get("SCINPVEL", None) * kms
        objra = primary_header.get("OBJRA", None) * u.deg
        objdec = primary_header.get("OBJDEC", None) * u.deg
        
        redshift = scinpvel / const.c.to('km/s') # redshift z = stellar velocity / speed of light
        distance = cosmo.luminosity_distance(redshift).to(u.Mpc) # luminosity distance

        # Doppler shift for depth
        dshift_map = hdul['EMLINE_GVEL'].data[0] * kms + scinpvel
        dshift_data = dshift_map.flatten()
        dshift_max = np.nanmax(np.abs(dshift_data))
        depth = (dshift_data / dshift_max) * depth_scale_factor

        # Flux data (from various emission lines)
        emline_data = hdul['EMLINE_GFLUX'].data + hdul['SPX_MFLUX'].data
        emline_data_masked = np.where(emline_data == 0, np.nan, emline_data)
        flux_data = emline_data_masked.reshape(35, -1).T

        # Colors: Use flux to map to color (based on the evolutionary stages)
        flux_norm = np.log1p(np.sum(flux_data, axis=1))  # Sum flux across selected channels
        colors = colormap(flux_norm)[..., :3]  # Apply colormap to flux values

        # Pixel coordinates in 3D space
        galaxy_coords = SkyCoord(ra=objra, dec=objdec, distance=distance, frame='icrs')
        galaxy_x = galaxy_coords.cartesian.x.to(u.Mpc).value
        galaxy_y = galaxy_coords.cartesian.y.to(u.Mpc).value
        galaxy_z = galaxy_coords.cartesian.z.to(u.Mpc).value
        
        coords = np.array([galaxy_x, galaxy_y, galaxy_z])

        # Apply PCA for grouping galaxies into clusters based on their location
        pca = PCA(n_components=3)
        pca.fit(coords)
        pca_coords = pca.transform(coords)

        # Clustering galaxies (using DBSCAN for dynamic grouping)
        clustering = DBSCAN(eps=0.5, min_samples=5).fit(pca_coords)
        labels = clustering.labels_

        # Prepare points for the 3D visualization
        points = np.column_stack([galaxy_x, galaxy_y, galaxy_z])
        colors = colors[labels == 0]  # Color filter based on DBSCAN cluster labels

        # Create the point cloud with Open3D
        pcd = o3d.geometry.PointCloud()
        pcd.points = o3d.utility.Vector3dVector(points)
        pcd.colors = o3d.utility.Vector3dVector(colors)

        return pcd

# Select a few files to visualize
selected_files = maps_files[:6]

# Generate point clouds for selected files
point_clouds = []
for file_path in selected_files:
    point_cloud = generate_point_cloud(file_path)
    point_clouds.append(point_cloud)

# Visualize with Open3D
o3d.visualization.draw_geometries(
    point_clouds,
    window_name="Galaxy Visualization",
    width=1200, height=800,
    left=50, top=50,
    point_show_normal=False,
    mesh_show_wireframe=False
)
```

```python
def get_info(file_path):
    kms = u.km / u.s
    
    with fits.open(file_path) as hdul:
        
        primary_header = hdul['PRIMARY'].header
        plateifu = primary_header.get("PLATEIFU", None)
        scinpvel = primary_header.get("SCINPVEL", None) * kms # inital object velocity
        objra = primary_header.get("OBJRA", None) * u.deg # right ascension
        objdec = primary_header.get("OBJDEC", None) * u.deg # declination
        
    redshift = scinpvel / const.c.to('km/s') # redshift z = stellar velocity / speed of light
    distance = cosmo.luminosity_distance(redshift).to(u.Mpc) # luminosity distance
    age = cosmo.lookback_time(redshift).to('Gyr') # lookback time of universe
    
    galaxy_coords = SkyCoord(ra=objra, dec=objdec, distance=np.abs(distance), frame='icrs')
    galaxy_x = galaxy_coords.cartesian.x.to(u.Mpc).value
    galaxy_y = galaxy_coords.cartesian.y.to(u.Mpc).value
    galaxy_z = galaxy_coords.cartesian.z.to(u.Mpc).value
    
    return plateifu, age.value, distance.value, redshift, scinpvel.value, [galaxy_x, galaxy_y, galaxy_z]


data_folder = '../../MaNGA/raw'
maps_files = [os.path.join(data_folder, f) for f in os.listdir(data_folder)]
with open('object_info.csv', 'w') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['Plateifu', 'Age', 'Distance', 'Redshift', 'Velocity', 'X', 'Y', 'Z'])
    
    for i, file_path in enumerate(maps_files, 1):
        plateifu, age, distance, redshift, scinpvel, coords = get_info(file_path)
        csvwriter.writerow([plateifu, age, distance, redshift, scinpvel, *coords])
        print(f'{i}/{len(maps_files)}', end='\r')
```
